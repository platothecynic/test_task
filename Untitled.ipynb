```python
# Створюємо копію дублікатів, щоб не змінювати оригінальний DataFrame.
temp = df[df.duplicated(keep='first')].copy()

# Додаємо стовпець, який показує, чи є рядок дублікатом.
temp['dupl'] = df.duplicated(keep='first')

# Вибираємо тільки рядки, які є дублікатами.
temp = temp[temp['dupl'] == True]

# Скидаємо індекси, щоб правильно працювати зі стовпцем index.
temp = temp.reset_index()

# Обчислюємо різницю між індексами сусідніх рядків.
temp['index_diff'] = temp['index'].shift(-1) - temp['index']

# Виводимо кількість рядків, у яких різниця між сусідніми індексами більше 1.
print(
    f'Кількість рядків, яка має різницю між сусідніми рядками більше 1 - {(temp.index_diff > 1).sum()} шт.'
)
```

```python
#Читаємо дані
df1 = pd.read_csv('table1.csv', parse_dates=['date_created'])
df2 = pd.read_csv('table2.csv', parse_dates=['date_created'])
```

```python
df1.head()
```

```python
#Видаляємо дублікати
df1 = df1[~df1.duplicated(keep='first')]
df2 = df2[~df2.duplicated(keep='first')]
```

```python
# Напишемо функцію для розрахунку вартості покупки в доларах залежно від типу покупки
def calculate_dollar_amount(row):
    if row['type_of_purchase'] == 'crystals':
        return row['spent'] * 5.5
    elif row['type_of_purchase'] == 'coins':
        return row['spent'] * 3.5
    else:
        return row['spent']  # Якщо тип покупки не відомий, просто повернемо вартість без змін

# Застосовуємо функцію до кожного рядка у вашому DataFrame
df2['spent_in_dollars'] = df2.apply(calculate_dollar_amount, axis=1)
```

```python
df2.spent.sum()
```

```python
#Для кожного користувача, додаємо значення дати його першого івенту
df1 = df1.merge(
    df1.groupby('id_user').agg(date_of_first_visit=('date_created',
                                                    'min')).reset_index(),
    how='left',
    left_on='id_user',
    right_on='id_user')
df1.head()
```

```python
#Залишаємо когорти котрі нас цікавлять
df1 = df1[(df1.date_of_first_visit >= "2022-12-12") & (df1.date_of_first_visit <= "2022-12-16")]
```

```python
#Перевіряємо чи немає перетинів в когортах і все правильно пораховано

# Задаємо два масиви id_user
array1 = df1[df1.date_of_first_visit == "2022-12-12"].id_user.unique()
array2 = df1[df1.date_of_first_visit == "2022-12-13"].id_user.unique()

# Знаходимо спільні значення
intersection = np.intersect1d(array1, array2)

# Кількість спільних значень
count_intersection = len(intersection)

print("Кількість спільних юзер айді в двох масивах:", count_intersection)
```

```python
#Додаємо стовпець lead_or_not, який вказує чи користувач мав покупку 
df1['lead_or_not'] = df1.id_user.isin(df2.id_user.unique())
```

```python
df1
```

```python
#Підраховуємо метрики, які потрібні до майбутнього аналізу
result_df = df1.groupby('date_of_first_visit').agg(
    un_users=('id_user', 'nunique'),
    reg_users=('id_user',
               lambda x: x[df1['type_of_event'] == 'reg'].nunique()))

result_df['conversion_to_reg'] = round(
    result_df['reg_users'] / result_df['un_users'] * 100, 2)

un_users_lead = df1[(
    df1.lead_or_not == True)].groupby('date_of_first_visit').agg(
        un_users_lead=('id_user', 'nunique')).reset_index()

result_df = result_df.merge(un_users_lead, how='left', on='date_of_first_visit')

result_df['conversion_reg_to_lead'] = round(
    result_df['un_users_lead'] / result_df['reg_users'] * 100, 2)

distinct_users_for_cohorts = df1[['date_of_first_visit', 'id_user']].drop_duplicates()

spents_by_users = df2.groupby('id_user').agg(spent=('spent', 'sum'),
                                     spent_dollars=('spent_in_dollars',
                                                    'sum')).reset_index()
spent_by_cohort = distinct_users_for_cohorts.merge(spents_by_users, how='left', on='id_user')


revenue = spent_by_cohort.groupby('date_of_first_visit').agg(coins__revenue=('spent', 'sum'),
                                                  dollars_revenue= ('spent_dollars', 'sum')).reset_index()

result_df = result_df.merge(revenue, how='left', on='date_of_first_visit')

result_df['ARPU'] = result_df['dollars_revenue']/result_df['un_users']
result_df['ARPPU'] = result_df['dollars_revenue']/result_df['un_users_lead']


result_df.set_index('date_of_first_visit', inplace=True)
result_df = result_df.round(2).astype(float)
result_df.T
```

Завдання 3.1 Напишіть SQL запит, що порахує ****середній за весь період конверт користувачів з тріального періоду в успішно сплачений другий платіж. Для розрахунку метрики включайте лише користувачів, які мають підписку tenwords_1w_9.99_offer.

```python
# Відкриття з'єднання
conn = sqlite3.connect('test.db')

# Виконання SQL-запиту
query = """
WITH filtered AS (
/*
Очищуємо таблицю від дублів, а також віж всіх транзакцій, користувачів
котрі мають product_id = 'tenwords_1w_9.99_offer' AND trial = True,
Оскільки Тріал не мав би попадатись в цьому пакеті.
*/
    SELECT 
    DISTINCT *
    FROM table1
    WHERE product_id = 'tenwords_1w_9.99_offer'
    AND user_id NOT IN (
                        SELECT DISTINCT user_id
                        FROM table1
                        WHERE product_id = 'tenwords_1w_9.99_offer'
                        AND trial = True
                        )
),
/*
Додаємо стовпець ranked і рангуємо для кожного унікального
користувача його транзакції за зростанням дати, перша транзакція
має 1 і так далі.
*/
ranked_data AS (
    SELECT 
    user_id, 
    purchase_date, 
    refunded,
    RANK() OVER(PARTITION BY user_id ORDER BY purchase_date ASC) as ranked
    FROM filtered
),
/*
Рахуємо кількість унікальних юзерів, в яких перша транзакція без рефаунда
тобто ranked = 1 AND refunded = 0
*/
first_purchase_users AS (
    SELECT 
    COUNT(DISTINCT CASE WHEN ranked = 1
                        AND refunded = 0
                        THEN USER_ID END) AS un_users_purchase_first_time      
    FROM ranked_data
),
/*
Рахуємо кількість унікальних юзерів, в яких перша транзакція без рефаунда
тобто ranked = 1 AND refunded = 0, а також друга транзакція успішна
ranked = 2 AND refunded = 0.
*/
second_purchase_users AS (
    SELECT
    COUNT(DISTINCT CASE WHEN ranked = 2
                        AND refunded = 0
                        THEN USER_ID END) AS un_users_purchase_second_time      
    FROM ranked_data
    WHERE user_id IN (
                    select DISTINCT user_id
                    FROM ranked_data
                    WHERE refunded = 0
                    and ranked = 1
                    )
)
SELECT
/*
Виводимо кількість унікальних користувачів які не зробили рефаунд при першій
транзакції, а також кількість користувачів котрі не зробили рефаун підчас
тругої транзакції, з умовою що перша була успішна,
та конверсію для таких користувачів, в другу успішну підписку.
*/
un_users_purchase_first_time,
un_users_purchase_second_time,
ROUND(
    CAST(un_users_purchase_second_time AS float)
    / un_users_purchase_first_time * 100
    ,2) AS conversion_to_sec_sub
FROM first_purchase_users, second_purchase_users
"""
result_df = pd.read_sql_query(query, conn)

# Закриття з'єднання
conn.close()
result_df
```

**Завдання 3.2**

Уявіть, що у вас є таблиця у базі даних, схема якої повторює схему таблиці представленої у **([файл 1](https://docs.google.com/spreadsheets/d/1px0EF_W5LUQa-x1W9ocuVvb0FrOG0_C_Qkdnyi3jS9g/edit?usp=sharing)).** Напишіть SQL запит, що пронумерує транзакції у таблиці відносно користувача за часом створення транзакції (purchase_date).

Для цього створіть поле, яке називатиметься transaction_number. Якщо певна транзакція була першою транзакцією користувача - її transaction_number буде 1, якщо другою - 2 і тд.

```python
# Відкриття з'єднання
conn = sqlite3.connect('test.db')

# Виконання SQL-запиту
query = """
WITH filtered AS (
/*
Очищуємо таблицю від дублів
*/
    SELECT 
    DISTINCT *
    FROM table1
    ORDER BY user_id, purchase_date ASC
)
/*
Додаємо стовпець transaction_number і рангуємо для кожного унікального
користувача його транзакції за зростанням дати, перша транзакція
має 1 і так далі.
*/
SELECT *
,RANK() OVER (PARTITION BY user_id ORDER BY purchase_date ASC) as transaction_number
FROM filtered
"""
result_df = pd.read_sql_query(query, conn)

# Закриття з'єднання
conn.close()
result_df.tail(20)
```

**Завдання 3.3**

Напишіть SQL запит, який поверне user_id тих користувачів, в яких перша транзакція (transaction_number = 1) має refunded=False, а друга - refunded=True.

```python
# Відкриття з'єднання
conn = sqlite3.connect('test.db')

# Виконання SQL-запиту
query = """
WITH filtered AS (
/*
Очищуємо таблицю від дублів
*/
    SELECT 
    DISTINCT *
    FROM table1
),
ranked_data AS (
/*
Додаємо стовпець transaction_number і рангуємо для кожного унікального
користувача його транзакції за зростанням дати, перша транзакція
має 1 і так далі.
*/
    SELECT 
    user_id, 
    purchase_date, 
    refunded,
    RANK() OVER(PARTITION BY user_id ORDER BY purchase_date ASC) as transaction_number
    FROM filtered
),
second_purchase_users AS (
/*
Для користувачів в яких перша транзакція була refund, виводимо user_id
якщо друга транзакція була успішною
*/
    SELECT
    CASE WHEN transaction_number = 2 AND refunded = 0 THEN user_id END AS user_id   
    FROM ranked_data
    WHERE user_id IN (
                    select DISTINCT user_id
                    FROM ranked_data
                    WHERE transaction_number = 1
                    AND refunded = 1
                    )
)
/*
Виводимо список user_id
*/
SELECT
user_id
FROM second_purchase_users
WHERE user_id IS NOT NULL
"""
result_df = pd.read_sql_query(query, conn)

# Закриття з'єднання
conn.close()
result_df
```
